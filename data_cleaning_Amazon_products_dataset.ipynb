```python
import pandas as pd
```

```python
pd.__version__
```

```python
data = pd.read_csv('data/amazon_product.csv', engine='pyarrow')
```

```python
data.head()
```

# Deleting redundant columns

```python
data.columns
```

```python
data.drop(columns= ['', 'asin', 'product_url', 'product_photo', 'product_num_offers', 'product_minimum_offer_price','is_amazon_choice', 'is_prime', 'climate_pledge_friendly', 'sales_volume', 'delivery', 'has_variations'], inplace=True)
```

```python
data
```

```python
data.columns
```

# Renaming the columns

```python
data_new_columns_names = []

for i in data.columns:
    data_new_columns_names.append(i.capitalize())
```

```python
data.columns = data_new_columns_names
```

```python
data.columns
```

# Dropping duplicates

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

# Remove the NaN values from dataset

```python
data.isna().sum()
```

```python
data.isna().sum().sum()
```

```python
data.shape
```

```python
data = data[['Product_title', 'Product_price', 'Currency', 'Product_star_rating', 'Product_num_ratings', 'Is_best_seller']]
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum().sum()
```

# Check for some more Transformations

```python
data.head(3)
```

```python
data['Product_price'] = data['Product_price'].str.replace('$','')
```

```python
data['Product_price'] = data['Product_price'].astype('float')
```

```python
data['Product_price'] = data['Product_price'].apply(lambda x: round(x))
```

```python
data.head(3)
```

```python
data.dtypes
```

```python
data.head()
```

```python
data['Is_best_seller'].unique()
```

```python
data['Is_best_seller'] = data['Is_best_seller'].apply(lambda x: 1 if x==True else 0)
```
