```python
data = pd.read_csv('data/netflix.csv', engine='pyarrow')
```

```python
data.head(20)
```

```python
data.info()
```

## Deleting redundant columns

```python
data.head()
```

```python
data.drop(columns= "rating", inplace=True)
```

```python
data.head()
```

## Renaming the columns

```python
data.columns
```

```python
new_columns_names = []

for i in data.columns:
    new_columns_names.append(i.capitalize())
```

```python
new_columns_names
```

```python
data.columns = new_columns_names
```

```python
data.head()
```

## Dropping duplicates

```python
data.duplicated().sum()
```

```python
# for cleaning duplicates

data.drop_duplicates(inplace=True)
```

## Remove the NaN values from dataset

```python
data.isna().sum()
```

```python
# if we had na values

data.dropna(inplace=True)
```

## Check for some more Transformations

```python
data['Show_id'].unique()
```

```python
data['Show_id'] = data['Show_id'].apply(lambda x: x.split("s")[1])
```

```python
type(data['Show_id'][1])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
data.dtypes
```

```python
data['Date_added'] = data['Date_added'].apply(lambda x: x.replace("/", "-"))
```

```python
data
```

```python
df = data['Country'].value_counts().reset_index()
```
