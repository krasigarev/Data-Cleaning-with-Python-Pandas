```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/FashionDataset.csv')
```

```python
data.head()
```

# Deleting redundant columns

```python
data.info()
```

```python
data.shape
```

```python
data.columns
```

```python
data.drop(columns= 'Unnamed: 0',inplace=True)
```

```python
data
```

# Renaming the columns

```python
data.rename(columns= {"Deatils": "Datails"}, inplace=True)
```

```python
data
```

# Dropping duplicates

```python
data.duplicated().value_counts().plot(kind='barh')
plt.title("Duplicated Rows")
plt.xlabel("Duplicate or not")
plt.ylabel("Amount of rows")
plt.legend(loc=1)
plt.show()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(columns= 'index', inplace=True)
```

```python
data.tail()
```

# Remove the NaN values from dataset

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data
```

# Check for some more Transformations

```python
data['BrandName'] = data['BrandName'].apply(lambda x: x.capitalize())
```

```python
data['Datails'] = data['Datails'].apply(lambda x: x.capitalize())
```

```python
data
```

```python
data['Sizes'] = data['Sizes'].str.replace('Nan', 'NA VALUED')
```

```python
data = data[data['Sizes'] != 'NA VALUED']
```

```python
data['Discount'] = data['Discount'].replace("Nan", "No Discount")
```

```python
data
```

```python
data = data.reset_index().drop(columns='index')
```

```python
data
```

# Export Data

```python
data.to_excel('data/cleaned_fashion_excel.xlsx')
```

```python
data.to_csv('data/cleaned_fashion_csv.csv', index=False)
```
