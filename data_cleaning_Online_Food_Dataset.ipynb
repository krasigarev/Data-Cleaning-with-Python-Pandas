# Deleting redundant columns

```python
data.head(1)
```

```python
data.columns
```

```python
columns_to_keep = ['Age', 'Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Family size', 'Output', 'Feedback']
columns_to_drop = ['latitude', 'longitude', 'Pin code', '']
```

```python
numbers_columns = len(columns_to_keep) + len(columns_to_drop)
```

```python
print(data.shape[1] == numbers_columns)
```

## 1st Method, filtering the columns that we wanted

```python
modify_data = data[columns_to_keep]
```

```python
modify_data
```

# 2nd way, dropping the unnecessary columns

```python
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data
```

# Renaming the columns

```python
data.columns
```

```python
data.rename(columns={'Monthly Income': 'Monthly Salary'}, inplace=True)
```

```python
data
```

# Dropping duplicates

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

# Remove the NaN values from dataset

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

# Check for some more Transformations

```python
data['Age (Monthd)'] = data['Age'] * 12
```

```python
data.head(3)
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'Female' else 0)
```

```python
data['Family size'] = data['Family size'].apply(lambda x: 'large' if x > 3 else 'small')
```

```python
data
```

```python
data['Output'] = data['Output'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
data['Output'] = data['Output'].astype(float)
```

```python
data.head(3)
```

# Export Data

```python
data.to_excel('data/cleaned_data_food_excel.xlsx', index=False)
```

```python
data.to_csv('data/cleaned_data_food_csv.csv', index=False)
```
