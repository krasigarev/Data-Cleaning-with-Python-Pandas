```python
import pandas as pd
import numpy as np
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
data = {
    "First Name": np.random.choice(["Alice", "Bob", "Charlie", "David"], 20),
    "Last Name": np.random.choice(["Smith", "Johnson", "Williams"], 20),
    "Age": np.random.randint(20, 50, 20),
    "Salary ($)": np.random.randint(40000, 90000, 20),
    "Start Date": pd.date_range(start='2020-01-01', periods=20, freq='M'),
    "Department": np.random.choice(['HR', 'IT', 'Finance', 'Marketing'], 20),
    "Employed?": np.random.choice([True, False], 20),
    "Notes": [np.nan if i % 5 == 0 else 'Note' for i in range(20)]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.rename(columns= {'Employed?': 'Employed'}, inplace=True)
```

```python
df['Employed'] = df['Employed'].apply(lambda x: 1 if x == True else 0)
```

```python
df.head()
```

```python
df['leval Salary'] = df['Salary ($)'].apply(lambda x: 'low leval' if x < 60000 else 'medium leval' if x < 80000 else 'high leval')
```

```python
df['leval Salary'].value_counts().plot(kind='pie', title='Top Leval Salary')
```

```python
df['Department'].value_counts().plot(kind='barh', title='Count Employees at Departments', xerr=0.4)
```
