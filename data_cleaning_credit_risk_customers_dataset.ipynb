```python
data = pd.read_csv('data/credit_customers.csv', engine='pyarrow')
```

```python
data.head(5)
```

# Deleting redundant columns

```python
data.columns
```

```python
data.shape
```

```python
data.drop(columns= ['checking_status', 'duration', 'credit_history', 'purpose', 'employment', 'installment_commitment', 
                    'personal_status', 'other_parties', 'residence_since', 'property_magnitude', 'other_payment_plans', 'num_dependents'], inplace=True)
```

```python
data
```

```python
data.drop(columns= 'savings_status', inplace=True)
```

```python
data.head(3)
```

# Renaming the columns

```python
data.columns
```

```python
new_columns_data_names = []

for i in data.columns:
    new_columns_data_names.append(i.capitalize())
```

```python
new_columns_data_names
```

```python
data.columns = new_columns_data_names
```

```python
data.columns
```

# Dropping duplicates

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

# Remove the NaN values from dataset

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data
```
